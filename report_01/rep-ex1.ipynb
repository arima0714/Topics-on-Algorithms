```python
def gcd(a, b):
    if(b == 0):
        return a
    return gcd(b,a%b)

gcd(111, 30)
```

```python
def mod(a0, a1):
    return (a0%a1)

# 拡張ユークリッドの互除法
# input: 二つの整数a, b
# output: d=gcd(a, b)とax+by = dを満たす整数x, y
def ex_euclid(a, b):
    if abs(a) < abs(b): # a, bの絶対値を比較(abs(a) = |a|, abs(b) = |b|)
        a, b = b, a     # |a| < |b| の場合は, a(割られる数)とb(割る数)を入れ替える
        change_flag = 1 # a, bの入れ替えがある場合とない場合とで、出力が変わるので、入れ替えがあったかどうかを記録
    else:
        change_flag = 0 # a, bの入れ替えがある場合とない場合とで、出力が変わるので、入れ替えがあったかどうかを記録

    a0, a1 = a, b
    #x0 -> prevx, y0 -> prevy
    x0, x1 = 1, 0
    y0, y1 = 0, 1
    while True:
        if a1 == 0:
            if change_flag == 0:
                return [a0, x0, y0] # a, bの入れ替えがない場合はそのまま出力
            else:
                return [a0, y0, x0] # a, bを入れ替えた場合はy0, x0を入れ替えて出力

        # 以下の "=" の後にコードを書いて関数を完成させてください。
        # qはa0をa1で割った時の商で, rはその時の剰余です。
        r = mod(a0, a1)
        q = (a0-r)//a1
        x0, x1 = x1, x0 - q * x1
        y0, y1 = y1, y0 - q * y1
        a0 = a1
        a1 = r
        
ex_euclid(111, 30)
```

```python
def BinaryAlgorithm(num1, num2):
    num_1 = num1
    num_2 = num2
    print(f"初期値 $B_1 = {int(num_1)}, B_2 = {int(num_2)}$\n")
    k = 0
    print(f"$B_1$ , $B_2$ のどちらかが奇数になるまで両方を$2$で割る")
    while(num_1%2 == 0 and num_2%2 == 0):
        k += 1
        num_1 /= 2
        num_2 /= 2
    w = 1
    print(f"その結果、 $B_1 = {int(num_1)}, B_2 = {int(num_2)}$ となった\n")
    while(w != 0):
        w = num_1 - num_2
        print(f"$w = B_1 - B_2$ より $w = {int(w)}$\n")
        if(w == 0):
            print(f"$w == {int(w)} $なのでループは終了。\n")
            break
        print(f"$w$が奇数になるまで$2$で割る。\n")
        while(w % 2 == 0):
            w /= 2
        print(f"$w = 2^s t$ とすると $t = {int(w)}$ となった。\n")
        if(w > 0):
            print(f"$t > 0$ なので $B_1$ に ${int(w)}$ を代入する。\n")
            num_1 = w
        else:
            print(f"$t <= 0$ なので $B_2$ に ${int(w)}$ を代入する。\n")
            num_2 = -w
        time.sleep(2)
    return(2**k * num_2)
```
